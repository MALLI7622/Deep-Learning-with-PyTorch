```python
import numpy as np
from matplotlib import pyplot as plt
def sinplot(flip=1):
    
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)


sns.set_style("darkgrid", { 'font.sans-serif': 'Bitstream Vera Sans' })
sinplot()
sns.despine()
plt.title('Sine graph')
plt.show()
```
